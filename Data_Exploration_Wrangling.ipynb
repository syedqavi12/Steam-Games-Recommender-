Data from: https://cseweb.ucsd.edu/~jmcauley/datasets.html#google_local

```python
f = open('australian_user_reviews.json', 'rt', encoding="utf8")
val = []
for line in f:
    fields = eval(line)
    val.append(fields)
val
```

```python
dicti ={'user_id': [], 'item_id': [], 'review': [], 'recommend': []}
for user in val:
    for item in user['reviews']:
        dicti['user_id'].append(user['user_id'])
        dicti['item_id'].append(item['item_id'])
        dicti['review'].append(item['review'])
        dicti['recommend'].append(item['recommend'])
game_df = pd.DataFrame(dicti)
```

```python
game_df
```

```python
f = open('australian_users_items.json', 'rt', encoding="utf8")
val = []
for line in f:
    fields = eval(line)
    val.append(fields)
val[:1]
```

```python
dicti ={'user_id': [], 'item_id': [], 'item_name': [], 'playtime': []}
for user in val:
    for item in user['items']:
        dicti['user_id'].append(user['user_id'])
        dicti['item_id'].append(item['item_id'])
        dicti['item_name'].append(item['item_name'])
        dicti['playtime'].append(item['playtime_forever'])
steam_df = pd.DataFrame(dicti)
```

```python
steam_df
```

```python
steam_merge = pd.merge(game_df, steam_df)
```

```python
most_playtime = steam_merge.groupby('item_name')['playtime'].agg(np.sum).sort_values(ascending=False)
most_playtime
```

```python
from collections import Counter
m_count = Counter(steam_merge.user_id)
m_count.most_common()
```

```python
user_freqs = np.array(list(m_count.values()))
#majority of user rate very little
```

```python
plt.plot(sorted(user_freqs, reverse=True))
plt.ylabel('# of games')
plt.xlabel('number of ratings per user')
plt.show()
```

```python
g_count = Counter(steam_merge.item_id)
g_count.most_common()
```

```python
games_freqs = np.array(list(g_count.values()))
```

```python
plt.plot(sorted(games_freqs, reverse=True))
plt.ylabel('Games frequency')
plt.xlabel('review length per games')
plt.show()
```
